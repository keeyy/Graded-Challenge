### GROUPING AND AGGREGATING

We can grouping and aggregating the datas. There are many ways to aggregation, ex: sum(), min(), max(), mean(), std(), std(), etc.

For this case we wanna see the total of confirmed, deaths, and recovered

```python
Confirmed = df.groupby('Terkonfirmasi').agg(sum)

print(Confirmed)
```

When the confirmed data are 4 there are 0 data of deaths and 996 datas of recovered, when the confirmed data are 15375 there are 119 datas of deaths and 0 data of recovered, etc

```python
Deaths = df.groupby('Meninggal').agg(sum)

print(Deaths)
```

When the deaths data are 0 there are 5701 datas of confirmed and 1000 datas of recovered, when the deaths data are 1 there are 503 datas of confirmed and 0 data of covered, etc

```python
Recovered = df.groupby('Sembuh').agg(sum)

print(Recovered)
```

When the recovered data are 0 there are 11639350 datas of confirmed and 1831 datas of deaths, when the recovered data are 1 there are 68 datas of confirmed and 0 data of deaths, when the recovered data are 4 there are 980 datas of confirmed and 0 data of deaths
